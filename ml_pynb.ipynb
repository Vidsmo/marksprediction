```python
import pandas as pd
data ={
    'Hours':[1,2,3,4,5,6,7,8,9],
    'Scores':[45,50,55,70,80,88,89,95,98]
}
df = pd.DataFrame(data)
print(df)
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['Hours'],df['Scores'],color='pink')
plt.xlabel("hoursstudied")#what will be diplayed on x axis means kya naam hoga uska
plt.ylabel("score")
plt.show()
#display the data frame in form of scattered grraph
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
x = df[['Hours']]
y = df['Scores']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=12)
model = LinearRegression()
model.fit(x_train, y_train)
print("model trained succesfully")
```

```python
y_pred = model.predict(x_test)
y_pred = y_pred.clip(0, 100)

result = pd.DataFrame({
    'Hours': x_test['Hours'].values,
    'Actual Score': y_test.values,
    'Predicted Score': y_pred
})
print(result)
```
